### Null values imputation

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

**Categorical variables**

```python
# create categical null value list
cat_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 0 and df[feature].dtype == "O"]
cat_nan
```

```python
# percentage of cat null values in total data
for feature in cat_nan:
    print("{}: {} %" .format(feature, np.round(df[feature].isnull().mean()*100,2)))
```

```python
df.head()
```

```python
# Replacing missing values with a new label
def replace_cat_feature(dataset,feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna("Missing")
    return data
```

```python
df = replace_cat_feature(df, cat_nan)
df.head()
```

```python
df[cat_nan].isnull().sum()
```

```python
for feature in cat_nan:
    data= df.copy()
    print(feature," :", data[feature].unique())
```

**Continues variables**

```python
numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum()>0 and df[feature].dtype != "O"]
```

```python
numerical_with_nan
```

```python
df[numerical_with_nan].isnull().sum()
```

```python
data = df.copy()
for feature in numerical_with_nan:
    median = data[feature].median()
    data[feature].fillna(median, inplace = True)
    data[feature +"_Nan"] = np.where(data[feature].isnull(), 1, 0)
    
data[numerical_with_nan].isnull().sum()
```

```python
df = data
df.head()
```

### Encoding

```python
# Applying frequency encoding for categorical variables
cat_variables = df.select_dtypes("O").columns
cat_variables
```

**categorical variables**

```python
df[cat_variables]
```

```python
# replacing missing with 1 and other with zeros
data = df.copy()
for feature in cat_nan:   
    data[feature+"_Missing"] = (data[feature] == 'Missing').astype(int)
data.head()
```

```python
df = data
df.head()
```

```python
#df.drop("test", axis = 1, inplace = True)
```

```python
data = df.copy()
data.shape
```

```python
# Trying one hot encoding
data = pd.get_dummies(data = data, columns =cat_variables )
```

```python
data.shape
```

One hot encoding increases the number of features(diamensionality)

```python
# trying freaquency encoding
data = df.copy()
for feature in cat_variables:
    cat_freq = data.groupby(feature).size()/len(data)
    data[feature] = data[feature].map(cat_freq)
data.head()
```
